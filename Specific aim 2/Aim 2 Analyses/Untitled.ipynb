```python
import pandas as pd
import os

folder_path = 'all_data/updated_xls_format'
all_files = [f for f in os.listdir(folder_path) if f.endswith('.xlsx') or f.endswith('.xls')]

df_list = []

for file in all_files:
    file_path = os.path.join(folder_path, file)
    temp_df = pd.read_excel(file_path)
    df_list.append(temp_df)

# Concatenate all dataframes in the list into a single dataframe
df = pd.concat(df_list, ignore_index=True)
df.head
```

```python
df.shape
```

```python
# Converting the 'RECORDING_SESSION_LABEL' column to string type to handle the type discrepancy
df['RECORDING_SESSION_LABEL'] = df['RECORDING_SESSION_LABEL'].astype(str)

# Splitting the column again
df['participant'] = df['RECORDING_SESSION_LABEL'].apply(lambda x: x.split('_')[0] if '_' in x else x)
df['session'] = df['RECORDING_SESSION_LABEL'].apply(lambda x: x.split('_')[1] if '_' in x else '2')

df[['RECORDING_SESSION_LABEL', 'participant', 'session']].head(10)  # Displaying the relevant columns to verify the changes
```

```python
# Creating a new dataframe 'df2' by filtering out rows where 'condition' is equal to 'practice'
df2 = df[df['condition'] != 'practice']
#verify that we dropped rows
print(df.shape)
print(df2.shape)
```

```python
# Creating the 'aoi1' column based on the provided logic
df2['aoi1'] = df2.apply(lambda row: row['LEFT_IA_1_SAMPLE_COUNT'] if row['RIGHT_IA_1_SAMPLE_COUNT'] == '.' else row['RIGHT_IA_1_SAMPLE_COUNT'], axis=1)

df2[['RIGHT_IA_1_SAMPLE_COUNT', 'LEFT_IA_1_SAMPLE_COUNT', 'aoi1']].head()  # Displaying the relevant columns to verify the new variable
```

### create aoi2,3, and 4:

```python
# Creating 'aoi2', 'aoi3', and 'aoi4' columns based on the provided logic

df2['aoi2'] = df2.apply(lambda row: row['LEFT_IA_2_SAMPLE_COUNT'] if row['RIGHT_IA_2_SAMPLE_COUNT'] == '.' else row['RIGHT_IA_2_SAMPLE_COUNT'], axis=1)
df2['aoi3'] = df2.apply(lambda row: row['LEFT_IA_3_SAMPLE_COUNT'] if row['RIGHT_IA_3_SAMPLE_COUNT'] == '.' else row['RIGHT_IA_3_SAMPLE_COUNT'], axis=1)
df2['aoi4'] = df2.apply(lambda row: row['LEFT_IA_4_SAMPLE_COUNT'] if row['RIGHT_IA_4_SAMPLE_COUNT'] == '.' else row['RIGHT_IA_4_SAMPLE_COUNT'], axis=1)

df2[['RIGHT_IA_2_SAMPLE_COUNT', 'LEFT_IA_2_SAMPLE_COUNT', 'aoi2', 
     'RIGHT_IA_3_SAMPLE_COUNT', 'LEFT_IA_3_SAMPLE_COUNT', 'aoi3',
     'RIGHT_IA_4_SAMPLE_COUNT', 'LEFT_IA_4_SAMPLE_COUNT', 'aoi4']].head()  # Displaying the relevant columns to verify the new variables
```

```python
# Adjusted function to map AOIs to the correct location names
def map_aoi_to_correct_location(row):
    if row['aoi_label'] == 'aoi1':
        if row['target_location'] == 1:
            return 'target_location'
        elif row['phonemic_location'] == 1:
            return 'phonemic_location'
        elif row['semantic_location'] == 1:
            return 'semantic_location'
        elif row['unrelated_location'] == 1:
            return 'unrelated_location'
    elif row['aoi_label'] == 'aoi2':
        if row['target_location'] == 2:
            return 'target_location'
        elif row['phonemic_location'] == 2:
            return 'phonemic_location'
        elif row['semantic_location'] == 2:
            return 'semantic_location'
        elif row['unrelated_location'] == 2:
            return 'unrelated_location'
    elif row['aoi_label'] == 'aoi3':
        if row['target_location'] == 3:
            return 'target_location'
        elif row['phonemic_location'] == 3:
            return 'phonemic_location'
        elif row['semantic_location'] == 3:
            return 'semantic_location'
        elif row['unrelated_location'] == 3:
            return 'unrelated_location'
    elif row['aoi_label'] == 'aoi4':
        if row['target_location'] == 4:
            return 'target_location'
        elif row['phonemic_location'] == 4:
            return 'phonemic_location'
        elif row['semantic_location'] == 4:
            return 'semantic_location'
        elif row['unrelated_location'] == 4:
            return 'unrelated_location'

melted_df['location_name'] = melted_df.apply(map_aoi_to_correct_location, axis=1)

# Finalize the formatted DataFrame
formatted_gca = melted_df[['participant', 'targetword', 'session', 'aoi_label', 'location_name', 'count']]
formatted_gca.head(10)  # Displaying the first few rows of the updated data
```
